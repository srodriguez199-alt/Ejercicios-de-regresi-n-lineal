```python
# ejercicio1.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import shapiro, t, f as f_dist

# Datos
equipos = ['Leones','Águilas','Tigres','Toros','Pumas','Jaguares']
X = np.array([38,45,31,50,27,41], dtype=float)   # goles anotados
Y = np.array([22,28,39,33,36,29], dtype=float)   # goles recibidos
n = len(X)
alpha = 0.05

print("="*70)
print("ANÁLISIS DE REGRESIÓN LINEAL SIMPLE - EJERCICIO 1")
print("="*70)
print(f"Datos (n={n}):")
df = pd.DataFrame({'Equipo':equipos,'Goles_anotados':X,'Goles_recibidos':Y})
print(df.to_string(index=False))

# Estadísticos descriptivos
print("\n--- Estadísticos descriptivos ---")
print(df[['Goles_anotados','Goles_recibidos']].describe())

# Gráficos (mostrar al ejecutar)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); sns.histplot(X,kde=True); plt.title('Goles anotados (X)')
plt.subplot(1,2,2); sns.histplot(Y,kde=True); plt.title('Goles recibidos (Y)')
plt.tight_layout(); plt.show()

plt.figure(figsize=(6,5)); sns.scatterplot(x=X,y=Y,s=100)
plt.title('Diagrama de dispersión: Goles anotados vs Goles recibidos')
plt.xlabel('Goles anotados'); plt.ylabel('Goles recibidos'); plt.grid(True); plt.show()

# Normalidad
sh_x = shapiro(X); sh_y = shapiro(Y)
print("\n--- Pruebas de Normalidad (Shapiro-Wilk) ---")
print(f"Goles_anotados: W={sh_x.statistic:.4f}, p={sh_x.pvalue:.4f}")
print(f"Goles_recibidos: W={sh_y.statistic:.4f}, p={sh_y.pvalue:.4f}")

# Correlaciones
pearson_r, pearson_p = stats.pearsonr(X,Y)
spearman_r, spearman_p = stats.spearmanr(X,Y)
kendall_r, kendall_p = stats.kendalltau(X,Y)
print("\n--- Coeficientes de correlación ---")
print(f"Pearson r = {pearson_r:.4f}, p = {pearson_p:.4f}")
print(f"Spearman rho = {spearman_r:.4f}, p = {spearman_p:.4f}")
print(f"Kendall tau = {kendall_r:.4f}, p = {kendall_p:.4f}")

# Prueba de hipótesis para la correlación (dos colas)
r = pearson_r
gl = n - 2
t_calc = r * np.sqrt(gl) / np.sqrt(1 - r**2) if abs(r) < 1 else float('nan')
p_val = 2 * stats.t.sf(abs(t_calc), df=gl)
t_crit = t.ppf(1 - alpha/2, gl)
print("\n--- Prueba de hipótesis para correlación (H0: rho=0) ---")
print(f"t calculado = {t_calc:.4f}, gl = {gl}, t critico = ±{t_crit:.4f}, p = {p_val:.4f}")

# Regresión lineal
slope, intercept, r_value, p_value, stderr = stats.linregress(X,Y)
y_pred = intercept + slope*X
residuals = Y - y_pred
SSE = np.sum(residuals**2)
SST = np.sum((Y - Y.mean())**2)
SSR = SST - SSE
R2 = r_value**2
s_yx = np.sqrt(SSE/(n-2)) if n>2 else float('nan')
print("\n--- Modelo de regresión ---")
print(f"Ŷ = {intercept:.4f} + {slope:.4f} X")
print(f"R² = {R2:.4f}, SSE = {SSE:.4f}, SSR = {SSR:.4f}, SST = {SST:.4f}")
print(f"Error estándar de estimación s_yx = {s_yx:.4f}")

# Prueba de significancia de la pendiente
sb = stderr
t_slope = (slope - 0)/sb if sb!=0 else float('nan')
p_slope = stats.t.sf(abs(t_slope), df=gl)*2
t_crit_slope = t.ppf(1-alpha, gl)  # para una cola si quisieramos; aquí mostramos
print("\n--- Prueba de significancia de la pendiente ---")
print(f"t pendiente = {t_slope:.4f}, p (two-sided) = {p_slope:.4f}")

# ANOVA (F)
MSR = SSR/1
MSE = SSE/(n-2) if n>2 else float('nan')
F_calc = MSR / MSE if MSE!=0 else float('nan')
p_F = f_dist.sf(F_calc, 1, n-2) if n>2 else float('nan')
print("\n--- Tabla ANOVA (Regresión) ---")
print(f"MSR = {MSR:.4f}, MSE = {MSE:.4f}, F = {F_calc:.4f}, p = {p_F:.6f}")

# Intervalos de confianza y prediccion (ejemplo para x=mean and x extremes)
t_crit_int = t.ppf(1-alpha/2, gl)
sum_xx = np.sum((X - X.mean())**2)
def ic_pred(x):
    yhat = intercept + slope*x
    term = np.sqrt((1/n) + ((x - X.mean())**2) / sum_xx)
    margin = t_crit_int * s_yx * term
    return yhat - margin, yhat + margin
def ip_pred(x):
    yhat = intercept + slope*x
    term = np.sqrt(1 + (1/n) + ((x - X.mean())**2) / sum_xx)
    margin = t_crit_int * s_yx * term
    return yhat - margin, yhat + margin

print("\n--- Ejemplo de IC e IP ---")
for x0 in [X.min(), X.mean(), X.max()]:
    ic = ic_pred(x0); ip = ip_pred(x0)
    print(f"X={x0}: ŷ={intercept + slope*x0:.3f}, IC mean={ic}, IP individual={ip}")

# Gráficos de residuos
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.scatter(y_pred, residuals); plt.axhline(0,color='r',ls='--'); plt.title('Residuos vs ŷ'); plt.xlabel('ŷ'); plt.ylabel('residuos')
plt.subplot(1,2,2); stats.probplot(residuals, plot=plt); plt.title('Q-Q residuos')
plt.tight_layout(); plt.show()

sh_res = shapiro(residuals)
print("\n--- Prueba Shapiro-Wilk residuos ---")
print(f"W = {sh_res.statistic:.4f}, p = {sh_res.pvalue:.4f}")

# Identificar outliers
res_std = residuals / s_yx if s_yx!=0 else residuals*0
outliers = np.abs(res_std) > 2.5
print("\n--- Valores atípicos (residuos estandarizados > 2.5) ---")
for i,flag in enumerate(outliers):
    if flag: print(f"{equipos[i]}: residuo estandarizado = {res_std[i]:.4f}")

# ===========================
# RESUMEN EJECUTIVO (por partes, estilo codigo original)
# ===========================
print("\n" + "="*70)
print("RESUMEN EJECUTIVO")
print("="*70)
print("\n📊 MODELO DE REGRESIÓN:")
print(f"   Ecuación: Ŷ = {intercept:.4f} + {slope:.4f} × X")
print("\n   Interpretación:")
print(f"   • El punto de partida (b₀) es {intercept:.2f} goles recibidos cuando X=0 (interpretar con cautela).")
print(f"   • Por cada gol anotado adicional, se espera un cambio de {slope:.4f} goles recibidos (pendiente).")

print("\n📈 BONDAD DE AJUSTE:")
print(f"   • Coeficiente de correlación (r): {pearson_r:.4f}")
print(f"   • Coeficiente de determinación (R²): {R2:.4f}")
print(f"     └─ {R2*100:.2f}% de la variabilidad en goles recibidos es explicada por goles anotados.")

print("\n✅ PRUEBAS DE SIGNIFICANCIA (α = {0}):".format(alpha))
print(f"   1. Correlación (Pearson): t = {t_calc:.4f}, p = {p_val:.4f}")
print(f"   2. Pendiente: t = {t_slope:.4f}, p = {p_slope:.4f}")
print(f"   3. Modelo (ANOVA): F = {F_calc:.4f}, p = {p_F:.6f}")

print("\n🔍 VALIDACIÓN DE SUPUESTOS:")
print(f"   • Normalidad de residuos: {'CUMPLE' if sh_res.pvalue > alpha else 'NO CUMPLE'} (p = {sh_res.pvalue:.4f})")
print(f"   • Linealidad y homocedasticidad: revisar gráficos de residuos y dispersión")

print("\n💡 CONCLUSIÓN GENERAL:")
if p_slope < alpha and p_F < alpha:
    print("   ✓ EL MODELO ES VÁLIDO Y ÚTIL PARA PREDICCIONES")
    print("   El número de goles anotados parece estar relacionado con goles recibidos en estos datos (según las pruebas).")
else:
    print("   ⚠ EL MODELO TIENE LIMITACIONES")
    print("   No hay evidencia estadística fuerte de que la relación sea lineal y significativa con el nivel α especificado.")
print("="*70)
```
